**Data Reference:**

The file `data/dolphins.tsv` contains a representation of a social network dataset where dolphins have links between them if they frequently associated with one another.
- Taken from the Koblenz Network Collection by the University of Koblenz–Landau.
Dolphins network dataset – KONECT, April 2017.
http://konect.cc/networks/dolphins/.
    


**Use a for loop to populate a dict that will contain the neighbour list network representation**

````python
# Begin with edges
# Create an empty edge set
edges = set()

with open('data/dolphins.tsv', 'r') as file:
    for line in file:
        a, b = line.split('\t')
        e = (int(a), int(b))
        edges.add(e)
edges

````

**Verify that your solution matches that for nodes  55, 2, and  20.
There will be many nodes in the final dataset. 
The result should have the form:
{55: {2, 7, 8, 14, 20, 42, 58},
 2: {18, 20, 27, 28, 29, 37, 42, 55},
 20: {2, 8, 31, 55},...
 This shows that dolphin 55 is frequently associated with dolphins 2, 7, 8, 14, 20, 42, and 58**

````python
neighbors = {}  # Empty dictionary
for edge in edges:
    a, b = edge
    neighbors.setdefault(a, set()).add(b)
    neighbors.setdefault(b, set()).add(a)

# Sort neighbor lists
more_efficient = {k: sorted(v) for k, v in neighbors.items()}

print("{" + ','.join(f"{k}: {{{', '.join(map(str, v))}}}" for k,v in (more_efficient.items()))+"}")
````

With your entire neighbour list code in a single Jupyter notebook cell, use the timeit Magic command to report the performance of your code, executing 19 runs of 2000 loops each.

There should be no network data output from this cell.

````python
nodes_to_test = [55,2,20]
efficient_result = {node: more_efficient[node] for node in nodes_to_test}
````

````python
# Use %timeit to measure the performance
%timeit -n 2000 -r 19 efficient_result
````

**Discussion:**



The program utilises the setdefault() method of python dictionaries to efficiently which avoids the need for explicit checks like 'if a not in neighbors', making code more concise. Use of dictionary comprehension, f-strings and join() makes the code optimized.

**Design, implement, and verify the results of a second, less efficient implementation and record its equivalent performance metrics as above.**

````python
# Create an empty dictionary to store neighbor lists
neighbors_less_efficient = {}

for edge in edges:
    a, b = edge
    if a not in neighbors_less_efficient:
        neighbors_less_efficient[a] = set()
    if b not in neighbors_less_efficient:
        neighbors_less_efficient[b] = set()
    neighbors_less_efficient[a].add(b)
    neighbors_less_efficient[b].add(a)

less_efficient = {}
for k, v in neighbors_less_efficient.items():
    less_efficient[k] = sorted(v)
    
print("{" + ','.join(f"{k}: {{{', '.join(map(str, v))}}}" for k,v in (less_efficient.items())) + "}")
````

````python
less_eff_result = {node: less_efficient[node] for node in nodes_to_test}
````

````python
# Testing the performance of my less efficient solution
%timeit -n 2000 -r 19 less_eff_result
````

**Discussion:**



The code redundantly checks node existance, employs verbose sorting, and relies on inefficient string concatenation. Overall, it sacrifices simplicity and efficiency compared ot the more concise version. we can see the testing where efficient code took 16.9 nanoseconds where as less efficient code took 25.1ns. Also time for each loop iteration is 0.0326ns for efficient and 0.385ns for less efficient code. This indicates the efficient code is faster and more efficient.

#### 2.1.6. Reasoning 

In the previous step, we eliminated entries with missing data from the dataset.  If this dataset were to be used to train a machine learning model to predict vehicle prices, what is a potential drawback of this approach?  Describe an alternative approach and any related caveats of which we should be aware.



The potential drawback of this approach is the loss of potentially valuable information. By removing rows with missing values, we reduce the size of our dataset, which can lead to loss of diversity and important patterns in the data. This can result in a less accurate or less robust machine learning model. 

An alternative approach is imputation, where all the values are preserved by replacing missing data with an estimated values based on other available data/information. For numerical features like 'HP' and common  imputation method is to replace missing values with the mean, median or mode of the column. This allows us to retain all observations in the dataset while preserving the variability and patterns present in the data.

However, imputation has its caveats, it assumes that the missing values are missing at random, meaning that the probability if missingness depends only on observed data and not on unobserved data because of which it can introduce bias into the dataset or smooth out variability, impacting model performance. Thus, it is essential to carefully consider missing data mechanisms and choose appropriate imputation methods.

        - `active` status, i.e. whether they are eligible to compete in the next round of the tournament.
        - performance record (e.g., win/loss, score, number of cars they used in competition)
    - This class has some kind of `__str__` representation.
    - Be thoughtful about which of these characteristics are mutable and how to change them if they are not mutable.
$$$$
- There is a method to `buy_cars`.
    - This method will allow the `Team`s to each purchase their initial inventory.
$$$$
- There is a method to `_purchase_inventory`.
    - This internal method takes 1 argument: a single `Team` object.
    - With the information provided by the `Team`, isolate cars that are available from the **sponsoring `Make`r**.  
    - Considering the `Team` budget, select the set of cars representing the **optimal choice**.  That is, purchase as many cars as the budget permits, while trying to **maximise `MPG-H`**.  
        - For more comprehensive implementations of this combinatorial optimisation problem, expect this to be the **most computationally expensive** part of this sub-task.  It can reasonably require a few seconds to perform the optimal selection.
    - When these are selected, the `Team`'s budget and inventory should be updated.
$$$$    
- There is a method to `hold_event` (i.e., execute the tournament competition process).
    - Cycle through the pairwise matches, keeping track of `Team` performance metrics.
        - The teams will compete in a head to head match and either continue on or be **eliminated**.
    - After each match, allocate a **financial prize** to the winning `Team`.  You can decide how to implement this; perhaps the prize increases in every round, but it must be enough to allow them to take the next step...
    - After awarding the prize, allow the `Team` to `_purchase_inventory` again (increasing the number of cars in their inventory) before the next match.  
        - Newly purchased cars can be duplicates of members of the `Team`'s existing inventory, but only one any kind of car is permitted to be purchased in an individual `_purchase_inventory` action.
    - At the end of the tournament event, record the `Tournament.champion` `Team`.  
    

````python
import pandas as pd
import numpy as np
import random
import math

class Team:
    def __init__(self, sponsor, budget, inventory=None):
        #initialise a team instance with sponsor, budget, and optional inventory
        self.sponsor = sponsor
        self.budget = budget
        self.inventory = inventory if inventory else []
        self.active = True
        self.performance_record = {'wins': 0, 'loss': 0, 'score': 0}
        self.record = []

    def __str__(self):
        #string representation of the team instance
        return f"Team sponsored by {self.sponsor}, with ${self.budget} available, and {len(self.inventory)} cars."
    

class Tournament:
    def __init__(self, cars_df, name, num_teams=16):
        #initialise a tournament instance with DataFrame, name and num of teams 
        self.cars_df = cars_df.dropna()
        self.name = name
        self.num_teams = self._validate_teams(num_teams)
        self.teams = []
        self.sponsors = {}
        self.champion = None
        self.match = []
        self.win_records = {}
        self.data = self._prepare_data(cars_df)

    def _validate_teams(self, num_teams):
        #validate the number of teams provided
        if not isinstance(num_teams, int):
            raise ValueError("Number of teams must be an integer.")
        if num_teams <= 0:
            raise ValueError("Number of teams must be positive and non-zero.")
        if not math.sqrt(num_teams).is_integer():
            raise ValueError("Number of teams must be a perfect square.")
        return num_teams

    def _prepare_data(self, cars_df):
        #process the car data DataFrame
        
        cars_df['Price'] = round(cars_df['Price'] / 100) * 100
        cars_df = cars_df[cars_df['Year'] > 2000]
        maker_counts = cars_df['Make'].value_counts()
        
        popular_makers = maker_counts[maker_counts > 55].index
        cars_df = cars_df[cars_df['Make'].isin(popular_makers)]
        return cars_df

    def generate_sponsors(self, specified_sponsors=None):
        #Generate sponsors for the teams
        makers = list(self.data['Make'].unique())
        
        if specified_sponsors:
            num_specified = len(specified_sponsors)
            if num_specified > self.num_teams:
                raise ValueError("Number of specified sponsors cannot exceed the number of teams.")
            for sponsor in specified_sponsors:
                if sponsor not in makers:
                    raise ValueError(f"Invalid sponsor: {sponsor}.")
            self.sponsors = {f"Team_{i+1}": specified_sponsors[i] for i in range(num_specified)}
            remaining_teams = self.num_teams - num_specified
        else:
            remaining_teams = self.num_teams
       
        for i in range(remaining_teams):
            team_name = f"Team_{i+1 + num_specified}" if specified_sponsors else f"Team_{i+1}"
            sponsor = random.choice(makers)
            makers.remove(sponsor)
            self.sponsors[team_name] = sponsor

        lowest_prices = []
        highest_prices = []
        incr = 15000
      
        for sponsor in self.sponsors.values():
            sponsor_cars = self.data[self.data['Make'] == sponsor]
          
            min_price = min(sponsor_cars['Price'], default=float('inf'))
            max_price = max(sponsor_cars['Price'], default=float('inf'))
            lowest_prices.append(min_price)
            highest_prices.append(max_price)
      
        low = min(lowest_prices, default=float('inf'))
        high = max(highest_prices, default=float('inf'))
        self.budget = random.randint(low, high) // incr * incr

    def generate_teams(self):
        #generate team based on sponsors and budget
        for i, (team_name, sponsor) in enumerate(self.sponsors.items(), 1):
            self.teams.append(Team(sponsor, self.budget))

    def _purchase_inventory(self, team):
        #simulate the purchase of cars by a team within their budget
        maker_cars = self.cars_df[self.cars_df['Make'] == team.sponsor]
        affordable_cars = maker_cars[maker_cars['Price'] <= team.budget]
        affordable_cars = affordable_cars.sort_values(by='MPG-H', ascending=False)
        selected_cars = []
        budget_spent = 0
        for index, car in affordable_cars.iterrows():
            if budget_spent + car['Price'] <= team.budget:
                selected_cars.append(car)
                budget_spent += car['Price']
        team.inventory.extend(selected_cars)
        team.budget -= budget_spent

    def buy_cars(self):
        for team in self.teams:
            self._purchase_inventory(team)
            
    def hold_event(self):
        self.win_records = {team.sponsor: {'wins': 0, 'loss': 0, 'score': 0} for team in self.teams}
        match_teams = self.teams[:]
        r = 1
        while len(match_teams) > 1:
            winners = []
            losers = []
            for i in range(0, len(match_teams), 2):
                if i + 1 < len(match_teams) and match_teams[i].active and match_teams[i + 1].active:
                    team1, team2 = match_teams[i], match_teams[i + 1]
                    
                    #assign two teams participating in the match 
                    team1_score = sum(car['MPG-H'] for car in team1.inventory)
                    team2_score = sum(car['MPG-H'] for car in team2.inventory)
                    winner = team1 if team1_score > team2_score else team2
                    loser = team2 if team1_score > team2_score else team1
                   
                    #based on the scores calculate above, winner/loser is determined
                    winner.performance_record['wins'] += 1
                    loser.performance_record['loss'] += 1
                    
                    #update records
                    winners.append(winner)
                    winner.record.append('W')
                    loser.record.append('L')
                    winner.performance_record['score'] += team1_score + team2_score
                    
                    #marking losing team as inactive to eliminate
                    loser.active = False
                   
                    #prize amount added to the budget
                    prize_amount = 5000
                    winner.budget += prize_amount
                    self._purchase_inventory(winner)

            match_teams = winners
            losers.extend(loser for loser in match_teams if loser not in winners)
            r += 1
            
            #once there is only one team left, determined as champion
            self.champion = max(self.teams, key=lambda x: x.performance_record['wins'])

    def show_win_record(self):
        #generate and return a DataFrame representing win and loss records of each team
        win_record = {team.sponsor: {'Wins': team.performance_record['wins'], 'Losses': team.performance_record['loss']} for team in self.teams}
        return pd.DataFrame.from_dict(win_record, orient='index').rename(columns={'Wins': 'Wins', 'Losses': 'Losses'})

    
# execution
car = pd.read_csv('data/cardata_modified.csv')


````
